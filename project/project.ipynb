```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('weatherHistory.csv')

# Check for missing values
print("Missing Values Before Handling:")
print(df.isnull().sum())

# Convert 'Formatted Date' column to datetime format
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)

# Drop duplicates if any
df.drop_duplicates(inplace=True)

# Perform exploratory data analysis
summary_stats = df.describe()
correlation_matrix = df.corr()

# Feature engineering (create new features if needed)
# For example, extract hour/day/month from 'Formatted Date'
df['Hour'] = df['Formatted Date'].dt.hour
df['Day'] = df['Formatted Date'].dt.day
df['Month'] = df['Formatted Date'].dt.month

# Display the cleaned DataFrame
print("\nCleaned DataFrame:")
print(df.head())
```

```python

# Load and preprocess each agriculture-related CSV file
pesticides_df = pd.read_csv('pesticides.csv')
# Preprocess pesticides_df (Example: handle missing values by filling with 0)
pesticides_df.fillna(0, inplace=True)
# Other preprocessing steps for pesticides_df as needed

print(pesticides_df.isnull().sum())

# Fill missing values with appropriate methods
# For instance, filling NaN values in the 'Value' column with the mean
pesticides_df['Value'].fillna(pesticides_df['Value'].mean(), inplace=True)

# Check data types of columns and convert them if needed
print(pesticides_df.dtypes)

# If 'Year' column is not in datetime format, convert it
pesticides_df['Year'] = pd.to_datetime(pesticides_df['Year'], format='%Y')

# Data after preprocessing
print(pesticides_df.head())


```

```python

yield_df = pd.read_csv('yield.csv')


# Check for missing values
print(yield_df.isnull().sum())

# Check data types of columns
print(yield_df.dtypes)

# Convert 'Year' column to datetime format
yield_df['Year'] = pd.to_datetime(yield_df['Year'], format='%Y')

# Check the first few rows of the dataset
print(yield_df.head())

```

```python

# Load the Yield dataset
yield_2_df = pd.read_csv('yield_df.csv')

# Check for missing values
print(yield_2_df.isnull().sum())

# Check data types of columns
print(yield_2_df.dtypes)

# Convert 'Year' column to datetime format
yield_df['Year'] = pd.to_datetime(yield_2_df['Year'], format='%Y')

# Data after preprocessing
print(yield_2_df.head())
```

```python
# Concatenate along columns axis
agriculture_merged = pd.concat([pesticides_df, rainfall_df, temp_df, yield_df, yield_2_df], axis=1)
print(agriculture_merged.head(5))
```

```python
# Renaming columns in the agriculture_merged DataFrame
agriculture_merged.rename(columns={'Area': 'Area_agriculture', 'Year': 'Year_agriculture', 'Value': 'Value_agriculture', 'avg_temp': 'avg_temp_agriculture', 'average_rain_fall_mm_per_year': 'average_rainfall_agriculture'}, inplace=True)

# Renaming the 'Formatted Date' column in the weather DataFrame to match the column in the agriculture data
df.rename(columns={'Formatted Date': 'Year'}, inplace=True)

# Merging agriculture-related data with weather data based on the 'Year' column
final_merged = pd.merge(agriculture_merged, df, on='Year', how='inner')
```
